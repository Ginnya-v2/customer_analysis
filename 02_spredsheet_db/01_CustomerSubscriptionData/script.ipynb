```python
import pandas as pd

df_trn = pd.read_csv('../../01_src/01_CustomerSubscriptionData/.data/customer_cases.csv',index_col=0)
df_cs = pd.read_csv('../../01_src/01_CustomerSubscriptionData/.data/customer_info.csv',index_col=0)
df_prd_cs = pd.read_csv('../../01_src/01_CustomerSubscriptionData/.data/customer_product.csv',index_col=0)
df_prd = pd.read_csv('../../01_src/01_CustomerSubscriptionData/.data/product_info.csv',index_col=0)
```

```python
df = df_trn[df_trn['reason'] == 'signup'].drop(columns=['reason','channel'])
df = pd.merge(df, df_cs, how='left', left_on='customer_id', right_on='customer_id')
df = pd.merge(df, df_prd_cs, how='left', left_on='customer_id', right_on='customer_id')
df = pd.merge(df, df_prd, how='left', left_on='product', right_on='product_id')

df = df[df['date_time'] >= '2019-01-01']

df
```

```python
df['date_time'].str[:4].unique()  
```

```python
df.columns = [
  'ケースID', '日時', '顧客ID', '年齢', '性別', '商品', '契約日時', 'キャンセル日時', '商品名', '価格', '更新期間'
]
df
```

```python
df.to_csv('data.csv', index=False, encoding='utf-8-sig')
```
